```python
#importing libraries that I will use in this analysis
import pandas as pd
from soccerplots import radar_chart

#importing the queried table. I'm hiding the directory for privacy purposes.
df = pd.read_csv('D:\***\***\The Amateurs\Stats Sheet\CSV\comp.csv')
df
```

```python
#making sure everything's in its right type
df.dtypes
```

```python
#setting our player names as Index
df.set_index('Player', inplace=True)
df
```

```python
## For Trent Alexander-Arnold
## parameter names
params = df.columns.tolist()

## range values
ranges = [(0.15, 0.35), (1.00, 3.00), (1.50, 3.00), (0.50, 1.00), (10.0, 25.0), (5.00, 15.00),
          (2.50, 5.00), (0.0, 1.00), (30.0, 50.00), (5.00, 6.00),
          (40.00, 55.00), (0.70, 1.00)]

## parameter value
values = df.iloc[0]

## title values
title = dict(
    title_name='Trent Alexander-Arnold',
    title_color='#000000',
    subtitle_name='Liverpool FC',
    subtitle_color='#D00027',
    title_fontsize=18,
    subtitle_fontsize=15,
)

## plot radar chart
fig, ax = radar_chart.plot_radar(ranges=ranges, params=params, values=values, radar_color=['#D00027', '#FFFFFF'], title=title)
```

```python
## For Andrew Robertson

## setting the parameter values for Andrew Robertson
val_rob = df.iloc[1]

## title values
title_rob = dict(
    title_name='Andrew Robertson',
    title_color='#000000',
    subtitle_name='Liverpool FC',
    subtitle_color='#D00027',
    title_fontsize=18,
    subtitle_fontsize=15,
)

## plot radar chart
fig, ax = radar_chart.plot_radar(ranges=ranges, params=params, values=val_rob, radar_color=['#D00027', '#FFFFFF'], title=title_rob)
```
